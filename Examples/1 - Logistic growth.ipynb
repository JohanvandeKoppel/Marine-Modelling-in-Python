# The logistic growth equation

The logistic growth equation described the dynamics of a population $x$ with intrisic (maximal) growth rate $r$ and carrying capacity $K$. The logistic growth equation was first named in 1844–1845 by Pierre François Verhulst, who used it to describe population growth:

$$
\begin{aligned}
\frac{dN}{dt} & = r\big(1 - \frac{N}{K}\big)N \\
\end{aligned}
$$

The model was an important basis for Thomas Malthus, who argued that population cannot grow infinitively when restrainted by limited reseources, and will be limited by competition when density approaches the carrying capacity. Malthus in turn was a big inspiration to Charless Darwin, when he stated that organisms face a struggle for existence.

Below, we implement this model in Python.

### Import of important packages

```python
import numpy as np
from matplotlib import pyplot as plt
```

### Model parameters are defined here

```python
K = 100.0
R = np.linspace(0,K+10)
r = 0.2
```

```python
Growth=r*(1-R/K)*R
```

```python
fig=plt.plot(R,Growth)
plt.xlabel('Population size R')
Null=plt.ylabel('Population growth rate')
```

```python
EndTime=100
dT=1
```

The total number of iterations that the simulation has to do is determined by the required duration divided by the timestep. The int() function makes sure that the outcome is an discrete number.

```python
Iterations = int(EndTime/dT)
```

We define a function that calculates what the rate of change is for the population at a specific timestep as a function of its population size.

```python
def dR(R):
    return r*(1.0-R/K)*R
```

Now we can run the simulation, calculating the value for each next time step as a function of the value of the previous one.

```python
R=np.zeros(Iterations+1)
R[0]=1
for i in range(Iterations): 
    R[i+1] = R[i]+dT*dR(R[i])
```

After the simulation is finished, we can plot the results

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.plot(np.linspace(0,EndTime,len(R)),R)
a=plt.xlabel('Time')
b=plt.ylabel('Population size')
```
